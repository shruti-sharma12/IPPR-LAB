```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Function to display images using Matplotlib
def display_images(images, titles, rows, cols, figsize=(12, 6)):
    plt.figure(figsize=figsize)
    for i in range(len(images)):
        plt.subplot(rows, cols, i + 1)
        plt.imshow(images[i], cmap='gray')
        plt.title(titles[i])
        plt.axis('off')
    plt.show()

# Read the image
image = cv2.imread('image.jpg', cv2.IMREAD_GRAYSCALE)

# Apply Canny edge detection
edges = cv2.Canny(image, 50, 150)

# Apply Hough Transform to detect lines
lines = cv2.HoughLines(edges, 1, np.pi / 180, threshold=100)

# Plot the detected lines on a copy of the original image
image_lines = image.copy()

if lines is not None:
    for line in lines:
        rho, theta = line[0]

# Perform Hough Transform for circles
circles = cv2.HoughCircles(edges, cv2.HOUGH_GRADIENT, dp=1, minDist=50, param1=100, param2=30, minRadius=10, maxRadius=100)

# Draw the detected circles on a copy of the original image
image_circles = image.copy()
if circles is not None:
    circles = np.uint16(np.around(circles))
    for i in circles[0, :]:
        cv2.circle(image_circles, (i[0], i[1]), i[2], (0, 255, 0), 2)

# Display the original image, lines, and circles
display_images([image, edges, image_lines, image_circles],
               ['Original Image', 'Canny Edges', 'Detected Lines', 'Detected Circles'],
               1, 4)
```
