```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
image1 = cv2.imread('image1.jpeg')
image2 = cv2.imread('image2.jpeg')
```

```python
print("Image 1 shape:", image1.shape)
print("Image 2 shape:", image2.shape)
```

```python
height_of_image2, width_of_image2 = image2.shape[:2]
image1 = cv2.resize(image1, (width_of_image2, height_of_image2))
```

```python
result_addition = cv2.add(image2, image1)
```

```python
plt.imshow(cv2.cvtColor(result_addition, cv2.COLOR_BGR2RGB))
plt.title('Result after Addition')
plt.show()
```

```python
result_sub = cv2.subtract(image1, image2)
```

```python
plt.imshow(cv2.cvtColor(result_sub, cv2.COLOR_BGR2RGB))
plt.title('Result after Subtraction')
plt.show()
```

```python
image2 = cv2.resize(image2, (image1.shape[1], image1.shape[0]))

# Perform bitwise AND
bitwise_and = cv2.bitwise_and(image1, image2)


plt.subplot(133), plt.imshow(cv2.cvtColor(bitwise_and, cv2.COLOR_BGR2RGB)), plt.title('Bitwise AND')
plt.show()
```

```python
result_or = cv2.bitwise_or(image1, image2)
plt.subplot(132), plt.imshow(cv2.cvtColor(result_or, cv2.COLOR_BGR2RGB)), plt.title('Logical OR')
plt.show()
```

```python
result_xor = cv2.bitwise_xor(image1, image2)
plt.subplot(133), plt.imshow(cv2.cvtColor(result_xor, cv2.COLOR_BGR2RGB)), plt.title('Logical XOR')
plt.show()
```

```python
result_not = cv2.bitwise_not(image1)
cv2_imshow(result_not)
```

```python
height, width = image1.shape[:2]
# Define rotation angle and translation values
angle = 45
scale = 1.0
tx, ty = 50, 50
```

```python
rotation_matrix = cv2.getRotationMatrix2D((width / 2, height / 2), angle, scale)
rotated_image = cv2.warpAffine(image1, rotation_matrix, (width, height))
```

```python
# Perform translation
translation_matrix = np.float32([[1, 0, tx], [0, 1, ty]])
translated_image = cv2.warpAffine(image1, translation_matrix, (width, height))
```

```python
# Display original, rotated, and translated images
plt.subplot(131), plt.imshow(cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)), plt.title('Original')
plt.subplot(132), plt.imshow(cv2.cvtColor(rotated_image, cv2.COLOR_BGR2RGB)), plt.title('Rotated')
plt.subplot(133), plt.imshow(cv2.cvtColor(translated_image, cv2.COLOR_BGR2RGB)), plt.title('Translated')
plt.show()
```

1. What do you mean by gray level

In digital image processing, a "gray level" refers to the intensity value assigned to each pixel in a grayscale image. These intensity values typically range from 0 to 255 in an 8-bit grayscale image, where 0 represents black and 255 represents white. The gray level indicates the brightness of a pixel, with higher values representing brighter shades and lower values representing darker shades of gray. In simpler terms, it's the measure of how light or dark a pixel appears in a grayscale image.


2. Write the expression to find the number of bits to store a digital image

The expression to find the number of bits required to store a digital image depends on several factors, including the color depth and the dimensions of the image. Here's a basic expression for calculating the number of bits needed to store a grayscale image:

Number of bits = Width × Height × Bit depth

Width is the width of the image in pixels.
Height is the height of the image in pixels.
Bit depth is the number of bits used to represent each pixel's intensity value.
For example, if you have a grayscale image with dimensions 800 pixels wide by 600 pixels tall and a bit depth of 8 bits (common for grayscale images), the expression would be:

Number of bits = 800 × 600 × 8 = 4,800,000 bits

For color images, you would need to consider the number of bits per pixel for each color channel (usually red, green, and blue in RGB images) and possibly an additional channel for transparency (alpha channel). The expression would then be adjusted accordingly.

3. Name types of resolutions w. r. to a digital image.

Resolutions in the context of digital images can refer to various aspects of image quality and size. Here are some types of resolutions with respect to digital images:

1. Spatial Resolution: This refers to the number of pixels per unit length or area in an image. Higher spatial resolution indicates more detail in the image. It's commonly expressed as pixels per inch (PPI) or pixels per centimeter (PPCM).

2. Pixel Resolution: This is simply the total number of pixels in an image, often represented as width × height in pixels. For example, a 1920 × 1080 image has a pixel resolution of 2,073,600 pixels.

3. Color Resolution: Also known as bit depth or color depth, it refers to the number of bits used to represent the color of each pixel in the image. Common values are 8-bit (256 colors), 24-bit (16.7 million colors), and 32-bit (16.7 million colors with an additional alpha channel for transparency).

4. Print Resolution: This refers to the number of pixels per inch (PPI) required for high-quality printing. Typically, higher print resolutions result in sharper and more detailed prints.

5. Display Resolution: This is the number of pixels displayed on a screen, commonly expressed as width × height in pixels. Examples include HD (1280 × 720 pixels), Full HD (1920 × 1080 pixels), and 4K (3840 × 2160 pixels).

6. Optical Resolution: In scanning devices such as scanners or digital cameras, this refers to the physical resolution capability of the device's optical system to capture detail.

Understanding these different types of resolutions is important for ensuring the quality and compatibility of digital images across various devices and applications.


4. Specify the elements of the DIP system

In the context of Digital Image Processing (DIP), a typical DIP system comprises several essential elements to process digital images effectively. These elements include:

1. Image Acquisition: This is the process of capturing digital images from various sources such as digital cameras, scanners, satellite sensors, or medical imaging devices.

2. Preprocessing: Preprocessing involves enhancing and correcting digital images to improve their quality and prepare them for further processing. Common preprocessing techniques include noise reduction, contrast enhancement, and image resizing.

3. Image Enhancement: Image enhancement techniques are used to improve the visual quality of digital images by emphasizing certain features or removing unwanted elements. Examples include sharpening, histogram equalization, and filtering.

4. Image Restoration: Image restoration techniques are employed to recover degraded or corrupted images by reversing the effects of factors such as blur, noise, or compression artifacts.

5. Color Image Processing: For systems dealing with color images, specific processing techniques are used to manipulate color information effectively. This includes color space conversions, color correction, and color segmentation.

6. Image Compression: Image compression techniques are utilized to reduce the storage space required for digital images while minimizing the loss of visual quality. Compression can be lossless or lossy, depending on the application requirements.

7. Feature Extraction: Feature extraction involves identifying and extracting meaningful information or features from digital images, which can be used for further analysis or recognition tasks. Examples include edge detection, texture analysis, and object detection.

8. Image Analysis: Image analysis techniques are used to interpret and extract useful information from digital images for tasks such as object recognition, pattern recognition, and image classification.

9. Image Understanding: Image understanding refers to the process of interpreting and making sense of the content and context of digital images, often involving higher-level reasoning and decision-making algorithms.

10. Postprocessing and Visualization:Postprocessing involves applying final adjustments or transformations to processed images before visualization or output. Visualization techniques are used to display and interpret the results of image processing tasks effectively.

These elements work together in a Digital Image Processing system to manipulate, analyze, and interpret digital images for various applications across fields such as medicine, remote sensing, surveillance, and entertainment.

5. Write any four applications of DIP

Digital Image Processing (DIP) finds numerous applications across various fields due to its capability to manipulate and analyze digital images effectively. Here are four notable applications:

1. Medical Imaging: DIP plays a crucial role in medical imaging for tasks such as diagnosis, treatment planning, and image-guided interventions. It is used in techniques like X-ray imaging, MRI (Magnetic Resonance Imaging), CT (Computed Tomography), ultrasound, and PET (Positron Emission Tomography) scans. DIP algorithms help enhance image quality, remove noise, segment organs or tissues, detect anomalies, and track changes over time, aiding medical professionals in accurate diagnosis and treatment.

2. Remote Sensing: DIP is extensively used in satellite and aerial imaging for applications such as environmental monitoring, agriculture, urban planning, disaster management, and natural resource exploration. DIP techniques are employed to analyze satellite images to monitor land cover changes, track deforestation, assess crop health, detect environmental pollution, map urban growth, and identify geological features, among other tasks.

3. Biometrics: DIP plays a vital role in biometric systems for automated identification and verification of individuals based on their physiological or behavioral characteristics. DIP algorithms are used in fingerprint recognition, iris recognition, facial recognition, palmprint recognition, and voice recognition systems. These systems find applications in security systems, access control, forensic identification, border control, and mobile device authentication.

4. Digital Forensics: DIP techniques are utilized in digital forensics to analyze and extract evidence from digital images, videos, and other multimedia data for investigations related to cybercrime, fraud, intellectual property theft, and security breaches. DIP tools help forensic analysts recover deleted or obscured images, enhance low-quality images or videos, identify forged or tampered content, and analyze metadata to establish the authenticity and integrity of digital evidence.

These applications highlight the diverse range of fields where Digital Image Processing plays a crucial role, contributing to advancements in technology, science, healthcare, security, and environmental monitoring.
